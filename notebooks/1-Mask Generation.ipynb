```python
import os
import itertools
import matplotlib
import matplotlib.pyplot as plt

# Change to root path
if os.path.basename(os.getcwd()) != 'PConv2DKeras':
    os.chdir('..')
    
from utils.util import MaskGenerator

%matplotlib inline
```

```python
# generating the masks
mask_generator = MaskGenerator(512, 512, 3, rand_seed=42)

# plot the result
_, axes = plt.subplots(5, 5, figsize=(20, 20))
axes = list(itertools.chain.from_iterable(axes))

for i in range(len(axes)):
    img = mask_generator.sample()
    
    # plot the image on axes
    axes[i].imshow(img*255)
```
